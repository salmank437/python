```python
a = numpy.array((-2,1))
b = numpy.array((1,-3))

origin = numpy.array((0,0))

vectors = [a, b, a+b]
tails = [origin, a, origin]

plot_vector(vectors, tails)
```

```python
# basis vector

i = numpy.array((1,0))
j = numpy.array((0,1))

vec = 3*i + 2*j

vectors = [i, j, 3*i, 2*j, vec]
plot_vector(vectors)

pyplot.title("damnnnn");
```
