# Recreation of Bundle Motion Analysis for "Case 1" and "Case 2"

### Set-Up

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Y_1 = np.loadtxt('Data_Case_1.txt')
Y_2 = np.loadtxt('Data_Case_2.txt')
Y_3 = np.loadtxt('Data_Case_3.txt')

N = len(Y_1)
```

### Correlation Matrices R
$$
R = \frac{1}{N} Y^T Y
$$

```python
R_1 = (1/N) * (Y_1.T @ Y_1)
R_1 = R_1.astype(float)

R_2 = (1/N) * (Y_2.T @ Y_2)
R_2 = R_2.astype(float)

R_3 = (1/N) * (Y_3.T @ Y_3)
R_3 = R_3.astype(float)
```

### Eigenvectors and Values

```python
eig_vals1, eig_vecs1 = np.linalg.eig(R_1)

eig_vals2, eig_vecs2 = np.linalg.eig(R_2)

eig_vals3, eig_vecs3 = np.linalg.eig(R_3)
```
