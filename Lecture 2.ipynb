```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create the matrix B
B = np.array([[101, -1]])

# Step 2: Visualize the effect of matrix B on unit vectors and a grid of lines
unit_vector = np.array([[1, 0]])
grid_lines = np.array([[1, 0], [0, 1]])

transformed_unit_vector = np.dot(unit_vector, B)
transformed_grid_lines = np.dot(grid_lines, B)

# Step 3: Compute the inverse of matrix B
B_inverse = np.linalg.inv(B)

# Step 4: Visualize the effect of matrix B followed by its inverse
result_after_B_and_inverse = np.dot(np.dot(unit_vector, B), B_inverse)

# Step 5: Visualize the effect of the inverse matrix on its own
inverse_on_its_own = np.dot(unit_vector, B_inverse)

# Create subplots to visualize transformations
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot unit vectors
axs[0, 0].quiver(0, 0, unit_vector[0, 0], unit_vector[0, 1], angles='xy', scale_units='xy', scale=1, color='r', label='Unit Vector')
axs[0, 0].set_xlim(-5, 5)
axs[0, 0].set_ylim(-5, 5)
axs[0, 0].set_title('Unit Vector')

# Plot grid lines
axs[0, 1].quiver(0, 0, grid_lines[:, 0], grid_lines[:, 1], angles='xy', scale_units='xy', scale=1, color='b', label='Grid Lines')
axs[0, 1].set_xlim(-5, 5)
axs[0, 1].set_ylim(-5, 5)
axs[0, 1].set_title('Grid Lines')

# Plot the effect of B on unit vectors
axs[1, 0].quiver(0, 0, transformed_unit_vector[0, 0], transformed_unit_vector[0, 1], angles='xy', scale_units='xy', scale=1, color='g', label='Transformed Unit Vector')
axs[1, 0].set_xlim(-5, 5)
axs[1, 0].set_ylim(-5, 5)
axs[1, 0].set_title('B * Unit Vector')

# Plot the effect of B followed by its inverse
axs[1, 1].quiver(0, 0, result_after_B_and_inverse[0, 0], result_after_B_and_inverse[0, 1], angles='xy', scale_units='xy', scale=1, color='m', label='B * Unit Vector * B^-1')
axs[1, 1].set_xlim(-5, 5)
axs[1, 1].set_ylim(-5, 5)
axs[1, 1].set_title('B * Unit Vector * B^-1')

# Set labels
for ax in axs.flat:
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.legend()

# Display the plots
plt.tight_layout()
plt.show()
```
